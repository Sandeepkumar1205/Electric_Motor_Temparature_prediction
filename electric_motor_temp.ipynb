preprocessing

```python
df.dtypes
```

```python
df=df.drop(columns=["profile_id","torque"])
```

```python
x=df[
    [
        "ambient",
        "coolant",
        "u_d",
        "u_q",
        "motor_speed",
        "i_d",
        "i_q",
        "stator_yoke",
        "stator_winding"
        

    ]
]
y=df["pm"]
```

```python
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(x)
pickle.dump(scaler,open("transform.save","wb"))
```

Training

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, train_size=0.2, random_state=42)
```

```python
print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)
```

evaluation

```python
def evaluate_model(model,name):
    model.fit(X_train,y_train)
    preds=model.predict(X_test)
    rmse=np.sqrt(mean_squared_error(y_test,preds))
    r2=r2_score(y_test,preds)
    print(f"{name} -> RMSE: {rmse:.4f},R2:{r2:.4f}")
    return model,rmse,r2
```

```python
lr_model,lr_rmse,lr_r2=evaluate_model(
    LinearRegression(),"LinearRegression"

)
```

```python
dt_model,dt_rmse,dt_r2=evaluate_model(
    DecisionTreeRegressor(random_state=42),
    "Decision Tree Regressor"
)
```

```python
rf_model,rf_rmse,rf_r2=evaluate_model(
    RandomForestRegressor(n_estimators=20,random_state=42,n_jobs=-1),
    "Random Forest Regressor"
)
```

```python
results=pd.DataFrame({
    "Model":["Linear","Decision Tree",'Random forest'],
    "RMSE":[lr_rmse,dt_rmse,rf_rmse],
    "R2 Score":[lr_r2,dt_r2,rf_r2]
})
results
```
